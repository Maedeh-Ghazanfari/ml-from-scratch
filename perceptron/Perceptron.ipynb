```python
learning_rate = 0.1
epoch_count = 100
errors_log = []

for i in range(epoch_count):
    errors = 0
    # Shuffle data in each epoch for better learning
    df = df.sample(frac=1).reset_index(drop=True)

    # Extract new x and y after each epoch
    x = df[['Feature1', 'Feature2', 'Feature3']].values
    y_label = df['Class_Label'].values

    # Predict using current weights
    z = np.dot(x, w) + b
    y_pred = [1 if val >= 0 else 0 for val in z]
    df['Predicted'] = y_pred

    # Update weights if prediction is wrong
    for index, row in df.iterrows():
        x_i = row[['Feature1', 'Feature2', 'Feature3']].values
        if row['Class_Label'] != row['Predicted']:
            errors += 1
            w = w + learning_rate * (row['Class_Label'] - row['Predicted']) * x_i
            b = b + learning_rate * (row['Class_Label'] - row['Predicted'])

    # Track errors per epoch
    errors_log.append(errors)
    if errors == 0:
        break
print(errors_log)
```

```python
# Final prediction on the full dataset
final_preds = [1 if val >= 0 else 0 for val in np.dot(x, w) + b]
accuracy = np.mean(final_preds == y_label)
print("Final Accuracy:", accuracy)
print(y_label,final_preds)
```
